**MÉTODO DE NEWTON**

```python
vetor = []
t = range(0,10)
#f(x) = x² - 64
import numpy as np
from matplotlib import pyplot as plt
def f(x):
  y = x**2-64
  return y
def d(x):
  d = 2*x
  return d
x = 1 #chute inicial
for n in range(1,11): #vai repetir 10 vezes
  x = x - f(x)/d(x)
  print(f"A raiz da função é {x}") #deve estar na linha do x para plotar ográfico
  vetor.append(x)
plt.plot(vetor,t)
```

```python
#vetores para o gráfico
vetor = [] #vetor
t = range(0,10) #intervalo

import numpy as np #Comando como logarítimos, etc
from matplotlib import pyplot as plt # para plotar o gráfico
def f(x): #Função
  y = x**2+2*x
  return y #para reconhecer a função
  #derivada da função
def d(x):
  z = (2*x)+2
  return z
x = 4 #chute inicial
for n in range (0,10):#número de vezes que irá repetir o processo
  x = x - (f(x)/d(x))
  print(x)
#plotando o gráfico da função
  vetor.append(x)
plt.plot(vetor,t, color="red") #plota o gráfico (x,y) --> (vetor,t)
plt.grid() #aparece as grades
plt.xlabel("valores de x",color="blue")
plt.ylabel("valores de y")
plt.title("Newton-Rap")
```
